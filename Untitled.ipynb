```python
import pandas as pd
import os
import datetime as dt
```

```python
SOURCE_DIR = './03_06_IS_HOT_TOMTOM/'
TRAIN_DIR = './03_06_IS_HOT_TOMTOM2/'
```

```python
def velocity_to_los(x):
    velocity = x.tomtom_velocity
    if velocity <= 5:
        return 'E'
    elif 5 < velocity <= 15:
        return 'D'
    elif 15 < velocity <= 25:
        return 'C'
    elif 25 < velocity <= 35:
        return 'B'
    else:
        return 'A'
```

```python
def get_duration_with_LOS(x, df_data):
    try:
        temp_data = df_data.loc[(df_data.segment_id == x.segment_id)&(df_data.is_morning == x.is_morning)].sort_values(by=['period'], ascending = False).iloc[-1]
        if len(temp_data) > 0:
            if temp_data.base_LOS == x.base_LOS:
                if x.segment_id == 23795:
                    print(temp_data.period)
                result_duration = int(temp_data.duration_LOS) + 1
            else:
                result_duration = 1
        else:
            result_duration = 1
    except:
        result_duration = 1
    return result_duration

def get_duration_with_velocity(x, df_data):
    try:
        temp_data = df_data.loc[(df_data.segment_id == x.segment_id)&(df_data.is_morning == x.is_morning)].sort_values(by=['period'], ascending = False).iloc[-1]
        if len(temp_data) > 0:
            if temp_data.tomtom_velocity == x.tomtom_velocity:
               
                result_duration = int(temp_data.duration_velocity) + 1
            else:
                result_duration = 1
        else:
            result_duration = 1
    except:
        result_duration = 1
    return result_duration
```

```python
checked_time_list = {
   "1": ["16:00:00", "19:00:00"],
   "2": ["7:00:00", "10:00:00"]
}
```

```python
def convert_str_to_time(time_str):
    return dt.datetime.strptime(time_str, '%H:%M:%S').time()
```

```python
def check_times(need_to_checked_time):
    global checked_time_list
    checked_time = False
    for key, value in checked_time_list.items():
        if convert_str_to_time(value[0]) <= need_to_checked_time <= convert_str_to_time(value[1]):
            checked_time = True

    return checked_time
```

```python
csv_file_path = '2023-04-20.csv'
df = pd.read_csv(os.path.join(SOURCE_DIR, csv_file_path), index_col=0)
```

```python
period_list = sorted(df.period.unique(), reverse=True)
period_list[0]
```

```python
for csv_file_path in os.listdir(SOURCE_DIR):
    df = pd.read_csv(os.path.join(SOURCE_DIR, csv_file_path), index_col=0)
    period_list = sorted(df.period.unique(), reverse=True)
    df["base_LOS"] = df.apply(velocity_to_los, axis=1)
    result_df = pd.DataFrame()
    for index, period_item in enumerate(period_list):
        split_period = period_item.split("_")
        str_time = f"{split_period[1]}:{split_period[2]}:00"
        convert_time = convert_str_to_time(str_time)
        if not check_times(convert_time):
            continue
        temp_df = df.loc[df.period == period_item]
        if result_df.shape[0] == 0:
            temp_df["duration_velocity"] = [1]*temp_df.shape[0]
            temp_df["duration_LOS"] = [1]*temp_df.shape[0]
            result_df = temp_df
            continue

        else:
            temp_period_df = df.loc[df.period == period_list[index -1]]
            duration_velocity_list = temp_df.apply(get_duration_with_velocity, axis=1, args=(result_df,))
            duration_LOS_list = temp_df.apply(get_duration_with_LOS, axis=1, args=(result_df,))
            temp_df['duration_velocity'] = duration_velocity_list
            temp_df['duration_LOS'] = duration_LOS_list
            result_df = pd.concat([result_df, temp_df])
    result_df.to_csv(os.path.join(TRAIN_DIR, csv_file_path))
        
    
    
```
