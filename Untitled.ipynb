```python
import folium

# Tọa độ của điểm (ví dụ: Latitude: 21.0285, Longitude: 105.8542)
latitude = 10.830940288506234
longitude = 106.67745796558785

# Bán kính và màu sắc của hình tròn
radius = 20  # Bán kính (đơn vị: mét)
color = 'blue'  # Màu sắc

# Tạo một đối tượng bản đồ
map_osm = folium.Map(location=[latitude, longitude], zoom_start=14)

# Thêm một đánh dấu (circle marker) vào bản đồ với tọa độ, bán kính và màu sắc đã cho
folium.CircleMarker(
    [latitude, longitude],
    radius=radius,
    color=color,
    fill=True,
    fill_color=color,
    fill_opacity=0.6
).add_to(map_osm)

# Hiển thị bản đồ
map_osm.save('map.html')
```

```python
import folium

# Danh sách các điểm cần hiển thị
locations = [
    {"latitude": 21.0285, "longitude": 105.8542, "radius": 20, "color": "blue"},
    {"latitude": 21.0368, "longitude": 105.8342, "radius": 20, "color": "red"},
    {"latitude": 21.0179, "longitude": 105.8499, "radius": 20, "color": "green"}
]

# Tạo một đối tượng bản đồ
map_osm = folium.Map(location=[21.0285, 105.8542], zoom_start=14)

# Thêm các đánh dấu (circle marker) vào bản đồ cho mỗi điểm trong danh sách
for location in locations:
    folium.CircleMarker(
        [location["latitude"], location["longitude"]],
        radius=location["radius"],
        color=location["color"],
        fill=True,
        fill_color=location["color"],
        fill_opacity=0.6
    ).add_to(map_osm)

# Hiển thị bản đồ
map_osm.save('map.html')
```

```python
import folium

def highlight_route(start_point, end_point):
    # Tạo bản đồ OpenStreetMap
    m = folium.Map(location=start_point, zoom_start=14)

    # Tạo đoạn đường giữa hai điểm
    route = folium.PolyLine(locations=[start_point, end_point], color='red', weight=5)

    # Thêm đoạn đường vào bản đồ
    route.add_to(m)

    # Tạo điểm đầu và điểm cuối
    start_marker = folium.Marker(location=start_point, icon=folium.Icon(color='blue', icon='play'))
    end_marker = folium.Marker(location=end_point, icon=folium.Icon(color='green', icon='stop'))

    # Thêm điểm đầu và điểm cuối vào bản đồ
    start_marker.add_to(m)
    end_marker.add_to(m)

    # Highlight đoạn đường giữa hai điểm
    m.fit_bounds([start_point, end_point])

    # Hiển thị bản đồ
    return m

start_point = [21.0285, 105.8542]
end_point = [21.0368, 105.8491]

map_with_route = highlight_route(start_point, end_point)
map_with_route.save('map.html')
```
