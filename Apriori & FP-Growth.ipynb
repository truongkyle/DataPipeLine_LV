```python
# Import thư viện cần thiết
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpgrowth
import pandas as pd

# Chuẩn bị dữ liệu
dataset = [['bread', 'milk'], ['bread', 'diaper', 'beer', 'egg'], ['milk', 'diaper', 'beer', 'cola'], ['bread', 'milk', 'diaper', 'beer'], ['bread', 'milk', 'diaper', 'cola']]

# Sử dụng Apriori
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
print(te_ary)
print(te.columns_)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets_apriori = apriori(df, min_support=0.6, use_colnames=True)
print(frequent_itemsets_apriori)

# Sử dụng FP-Growth
frequent_itemsets_fpgrowth = fpgrowth(df, min_support=0.6, use_colnames=True)
print(frequent_itemsets_fpgrowth)
```
