```python
import requests
import json

def get_district_name(latitude, longitude):
    url = f"https://nominatim.openstreetmap.org/reverse?format=json&lat={latitude}&lon={longitude}"
    response = requests.get(url)
    data = response.json()
    print(data)
    if 'address' in data:
        if 'city' in data['address']:
            if data['address']['city'] == 'Thành phố Thủ Đức':
                return data['address']['city']
        if 'county' in data['address']:
            return data['address']['county']
        elif 'city_district' in data['address']:
            return data['address']['city_district']
        elif 'district' in data["address"]:
            return data['address']['district']
        elif 'town' in data['address']:
            return data['address']['town']
        elif 'village' in data['address']:
            return data['address']['village']
        

    return None

# Gọi hàm để lấy tên quận
latitude = 10.8013004  # Thay bằng tọa độ latitude của điểm bạn muốn kiểm tra
longitude = 106.711558  # Thay bằng tọa độ longitude của điểm bạn muốn kiểm tra
district_name = get_district_name(latitude, longitude)

if district_name:
    print(f"Đoạn đường này thuộc quận {district_name} của thành phố Hồ Chí Minh.")
else:
    print("Không tìm thấy thông tin về quận.")
```
